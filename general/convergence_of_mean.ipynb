# Observe how a mean of a distribution converges with samples

2025-03-22

Try on:
- gaussian
- cauchy

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

## Gaussian

```python
np.random.rand(13)
N = 200

x = np.random.randn(N) + 2.0
ns = np.arange(N) + 1
x_mean = np.cumsum(x) / ns
```

```python
plt.plot(ns, x_mean, '+-')
plt.grid()
```

```python
# larger scale
np.random.seed(43)

ns = [1, 2, 5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000, 20000, 50000, int(1e5), int(2e5), int(5e5)]
x_mean = []

for i, ni in enumerate(ns):
    res = (np.random.randn(ni) + 2.0).sum() / ni
    x_mean.append(res)
```

```python
plt.plot(ns, x_mean, '+-')
plt.grid()
plt.semilogx()
```

## Cauchy distribution

```python
%%time
np.random.seed(42)

ns = []
for o in range(8):
    tmp = [int(10**o), int(2*10**o), int(5*10**o)]
    ns.extend(tmp)
x_mean = []

for i, ni in enumerate(ns):
    res = (np.random.standard_cauchy(ni)).sum() / ni
    x_mean.append(res)
```

```python
plt.plot(ns, x_mean, '+-')
plt.grid()
plt.semilogx()
```
