```python
import numpy as np
from numpy.linalg import norm

import plotly.express as px
```

```python
np.random.seed(123)
```

```python
N = 3000

v = np.random.randn(N, 3)
v -= v.mean(0)
```

```python
v.shape
```

```python
px.histogram(v[:, 0])
```

```python
%%time
# compute mean mutual speed

vmut = []

for i in range(N):
    for j in range(i):
        vmut.append(norm(v[i] - v[j]))
        
np.mean(vmut)
```

```python
# verify that v_mut = sqrt(2) mean(v)
v_mean = np.sqrt(8.0 / np.pi)
coeff = np.mean(vmut) / v_mean

coeff, coeff / np.sqrt(2)
```
