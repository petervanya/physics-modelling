```python
import numpy as np
import pandas as pd
from scipy.integrate import odeint

import plotly.express as px
```

```python
import scipy
scipy.__version__
```

```python
def evolve(X, t, k1, k2):
    A, B, C, D = X
    dAdt = -k1 * A * B**2
    dBdt = -k1 * A * B**2
    dCdt = k1 * A * B**2 - k2 * C
    dDdt = k2 * C
    return [dAdt, dBdt, dCdt, dDdt]
```

```python
# initial condition
X0 = [1.5, 1.0, 0.0, 0.0]

# time vector
t = np.linspace(0, 10, 101)
k1 = 1.0
k2 = 1.0
```

```python
sol = odeint(evolve, X0, t, args=(k1, k2))
```

```python
df = pd.DataFrame(sol, columns=list('ABCD'), index=t)

px.line(df)
```
